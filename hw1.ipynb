```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

labor_data = pd.read_csv('labor.csv')
labor_data.head()

fig = plt.figure(figsize=(15, 7))
ax1 = fig.add_subplot(111)
plt.ylim(40, 70)
for (i, row) in labor_data.iterrows():
    plt.bar([i, i], [row['male_ave'], row['female_ave']],
            color=['#008AB8', '#CC6699'], width=0.8, align='center', edgecolor='none')

plt.xticks(np.arange(0,10), labor_data['year'], fontsize=11)
plt.xlabel('Year(B:male,R:female)')
plt.ylabel('% of Labor force')
;
```

（圖一）男女於96年至105年勞動力參與率

可以從圖中明顯看出，雖然女性百分比突破50%，卻仍然遠小於男性。


將婚姻狀態分為三種，分別為未婚、有配偶或同居及離婚分居或喪偶

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

labor_data = pd.read_csv('labor.csv')
labor_data.head()

fig = plt.figure(figsize=(15, 7))
ax1 = fig.add_subplot(111)
plt.ylim(40, 80)
for (i, row) in labor_data.iterrows():
    plt.bar([i], [row['male_ave']],color=['#008AB8'], width=0.8, align='center', edgecolor='none')
plt.plot(labor_data['male_unmarried'],'ro--',label='male_unmarried')
plt.plot(labor_data['male_married'],'b^--',label='male_married')
plt.plot(labor_data['male_divorce'],'ks--',label='male_divorce')
plt.legend(loc='upper right')

plt.xticks(np.arange(0,10), labor_data['year'], fontsize=11)
plt.xlabel('Year')
plt.ylabel('% of Labor force(Average)')
;
```

（圖二）男性勞動力平均參與率與三種婚姻狀態的比較圖

由圖二可知，男性已婚的勞動比例較高，於96年時與平均相差高達7%，而至105年為止差距已經越來越小。

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

labor_data = pd.read_csv('labor.csv')
labor_data.head()

fig = plt.figure(figsize=(15, 7))
ax1 = fig.add_subplot(111)
plt.ylim(25,70)
for (i, row) in labor_data.iterrows():
    plt.bar([i], [row['female_ave']],color=['#CC6699'], width=0.8, align='center', edgecolor='none')
plt.plot(labor_data['female_unmarried'],'ro--',label='female_unmarried')
plt.plot(labor_data['female_married'],'b^--',label='female_married')
plt.plot(labor_data['female_divorce'],'ks--',label='female_divorce')
plt.legend(loc='upper right')

plt.xticks(np.arange(0,10), labor_data['year'], fontsize=11)
plt.xlabel('Year')
plt.ylabel('% of Labor force')
```

（圖三）女性勞動力平均參與率與三種婚姻狀態的比較圖

由圖三可知，女性未婚的勞動比例較高，有日益增長的趨勢。

```python
labor_data = pd.read_csv('labor.csv')
labor_data.head()

fig = plt.figure(figsize=(15, 7))
ax1 = fig.add_subplot(111)
ax1.set_title('105year')
labels='M_Average','M_Unmarried','M_Married','M_Divorce','F_Average','F_Unmarried','F_Married','F_Divorce'
size=[labor_data.iloc[9, 1],labor_data.iloc[9, 2],labor_data.iloc[9, 3],labor_data.iloc[9, 4],
      labor_data.iloc[9,5],labor_data.iloc[9, 6],labor_data.iloc[9, 7],labor_data.iloc[9, 8]]
explode = (0, 0, 0, 0,0,0,0,0)
plt.pie(size, labels=labels, explode =explode ,autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')
plt.show()

```

（圖四）105年男女性各型態勞動力百分比圓餅圖

從上面四張圖可以看出，女性勞動比例雖然有在漸漸提升，始終和男性有很大的距離。
而男性已婚勞動參與比例比其他狀況高，女性則是未婚較高，可以推斷可能是傳統社會價值觀認為男性該負責賺錢養家，女性則負責在家中相夫教子的緣故。
